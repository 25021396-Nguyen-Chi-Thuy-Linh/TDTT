# W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. 
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
max_val = None
min_val = None
while True:
    n = int(input())
    if n == -1:
        break
    if max_val is None and min_val is None:
        max_val = n
        min_val = n
    else:
        if n>max_val:
            max_val = n
        if n<min_val:
            min_val = n
    print(n, end=' ')
if max_val is not None:
    print(f'so lon nhat la {max_val}\nso nho nhat la {min_val}')
```

# W4A26
[Fibonaci1] 
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ 
bằng tổng hai số trước đó 
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt 
quá A 

```python
A = int(input())
f0,f1 = 1, 1

if A < 1:
    print('Khong co so Fi nào thỏa mãn')
else:
    while True:
        f_next = f0 + f1
        if f_next > A:
            break
        
        f0 = f1
        f1 = f_next
    
    print(f'So Fi lon nhat la {f1}')
```

# W4A27
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có 
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng) 
Dùng cách: for trên index; KHÔNG dùng hàm split(), len() 

```python
chuoi = input()
len_chuoi = 0
for i in chuoi:
    len_chuoi += 1

so_tu = 0
trong_tu = False

for j in range(len_chuoi):
    if chuoi[j] != ' ':
        if not trong_tu:
            so_tu += 1
            trong_tu = True
    else:
        trong_tu = False

print(f'Chuoi "{chuoi}" co {so_tu} tu')
```

# W4A29
Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành 
thường và thường thành hoa. 

```python
chuoi = input()
chuoimoi = ''

do_dai = 0
for i in chuoi:
    do_dai += 1

for j in range(do_dai):
    if 'a' <= chuoi[j] <= 'z':
        chuoimoi += chuoi[j].upper()
    
    elif 'A' <= chuoi[j] <= 'Z':
        chuoimoi += chuoi[j].lower()
    else:
        chuoimoi += chuoi[j]

print(chuoimoi)
```

# W4A31
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in 
thường, bao nhiêu ký tự số

```python
chuoi = input()
inhoa = 0
inthuong = 0
so = 0
for i in chuoi:
    if 'A' <= i <= 'Z':
        inhoa += 1
    elif 'a' <= i <= 'z':
        inthuong += 1
    elif '0' <= i <= '9':
        so += 1
print(f'chuoi {chuoi} co {inhoa} ky tu in hoa, {inthuong} ky tu in thuong, {so} ky tu so')
```
