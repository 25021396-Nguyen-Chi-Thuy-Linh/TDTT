# W5A1
Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) -> 
int: 

```python
def Max_Of_Two(a,b):
    max_val = a
    if b>a:
        max_val = b
    return max_val
a,b = map(int, input().split())
print(Max_Of_Two(a,b))
```

# W5A2
Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]:  

```python
def swap(a,b):
    temp = a
    a=b
    b=temp
    return a,b
a,b = map(int, input().split())
print(swap(a,b))
```

# W5A3
Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def so_nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n%i == 0:
            return False
    return True

n = int(input())
print(so_nguyen_to(n))
```

# W5A4
VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def perfect_number(n):
    if n<6:
        return False
    s=0
    for i in range(1,n):
        if n%i == 0:
            s += i
            
    return s == n

n = int(input())
print(perfect_number(n))
```

# W5A7
Viết hàm giả lập một máy tính bỏ túi. 
Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat 
\) 
Biết các toán tử đơn giản thực hiện phép toán lần lượt là: 
Phép trừ \( - \) 
Phép cộng \( + \) 
Phép chia \( / \) 
Phép nhân \( * \) 
Kết quả làm tròn đến chữ số thập phân thứ hai. 
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)). 
Đầu ra: Kết quả phép toán. 

```python
def res(a, b, operat):
    if operat == '-':
        re = a-b
    if operat == '+':
        re = a+b
    if operat == '/':
        if b == 0:
            return 'Loi chia cho so 0'
        re = a/b
    if operat == '*':
        re = a*b
    return round(re, 2)
a,b = map(int, input().split())
operat = input()
print(res(a,b, operat))
```

# W5A8
Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng 
của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2. 
Cho hai số nguyên  x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. 
Khoảng cách Hamming là cái tên được đặt theo tên của Richard Hamming, người giới thiệu lý 
thuyết này trong tài liệu có tính cơ sở của ông về mã phát hiện lỗi và sửa lỗi (error-detecting and 
error-correcting codes). Nó được sử dụng trong kỹ thuật viễn thông để tính số lượng các bit trong 
một từ nhị phân (binary word) bị đổi ngược, như một hình thức để ước tính số lỗi xảy ra trong quá 
trình truyền thông, và vì thế, đôi khi, nó còn được gọi là khoảng cách tín hiệu (signal distance). Việc 
phân tích trọng số Hamming của các bit còn được sử dụng trong một số ngành, bao gồm lý thuyết 
tin học, lý thuyết mã hóa, và mật mã học. Tuy vậy, khi so sánh các dãy ký tự có chiều dài khác 
nhau, hay các dãy ký tự có xu hướng không chỉ bị thay thế không thôi, mà còn bị ảnh hưởng bởi dữ 
liệu bị lồng thêm vào, hoặc bị xóa đi, phương pháp đo lường phức tạp hơn, như khoảng cách 
Levenshtein (Levenshtein distance) là một phương pháp có tác dụng và thích hợp hơn. 

```python
def hamming_distance(x, y):
    xor_res = x^y
    dis = 0
    while xor_res > 0:
        xor_res = xor_res & (xor_res - 1)'
        dis += 1
    return dis
    
x, y = map(int, input().split())
print(hamming_distance(x,y))
```

# W5A9
Viết hàm trả về tổng các chữ số của số nguyên dương n 

```python
def tong_chu_so(n):
    s=0
    while n>0:
        s += n%10
        n //= 10
    return s
n = int(input())
print(tong_chu_so(n))
```

# W5A10
Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh 
xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự 
được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có 
thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành 
'z', 'b' thành 'b' và 'c' thành 'x'. 
Viết hàm  trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai 
từ đầu vào có cùng độ dài. 

```python
def dang_cau(a,b):
    if len(a) != len(b):
        return False
    
    mapping_a_to_b = {}
    mapped_b_char = set()
    
    for char_a, char_b in zip(a, b):
        if char_a in mapping_a_to_b:
            if mapping_a_to_b[char_a] != char_b:
                return False
            
        else:
            if char_b in mapped_b_char:
                return False
            
        mapping_a_to_b[char_a] = char_b
        mapped_b_char.add(char_b)
        
    return True

a = input()
b = input()
print(dang_cau(a,b))
```

# W5A12
Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 
0 <= |r| < |divisor|. 

```python
def divide_with_remainder(dividend, divisor):
    if divisor == 0:
        return 'Error'
    
    q = dividend // divisor
    r = dividend % divisor
    
    return q,r

dividend, divisor = map(int, input().split())
print(divide_with_remainder(dividend, divisor))
```

# W5A14
Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op. 

```python
def compute(a,b,op):
    if op == '+':
        res = a + b
    elif op == '-':
        res = a - b
    elif op == '*':
        res = a*b
    elif op == '/':
        if b == 0:
            return 'Error'
        res = a/b
        
    return res

a, b = map(int,input().split())
op = input()
print(compute(a,b,op))
```

# W5A15
Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.  

```python
def is_right_triangle(a,b,c):
    if a+b > c and a+c > b and b+c > a:
        if a**2 == b**2 + c**2 or b**2 == a**2 + c**2 or c**2 == a**2 + b**2:
            return True
    
    return False

a,b,c = map(int,input().split())
print(is_right_triangle(a,b,c))
```
