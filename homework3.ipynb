# Bài 14: 
Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập 
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
a=int(input())
if a==-1:
    print('Hãy nhập số khác -1')
else:
    max_a=a
    min_a=a
    while True:
        a = int(input('Nhập -1 để dừng'))
        if a == -1:
            break
        if a>max_a:
            max_a=a
        if a<min_a:
            min_a=a
        
print(f'Số lớn nhất là: {max_a}')
print(f'Số nhỏ nhất là: {min_a}')
```

# Bài 15: 
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều 
cách nhất có thể, ko dùng mảng.

```python
n = int(input())
d=1
if n<0:
    print('Hãy nhập số nguyên dương')
elif n<10:
    d=1
else:
    while n>=10:
        d+=1
        n//=10
print(d)
```

# Bài 16: 
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu 
chữ số lẻ

```python
n = int(input())
chan = 0
le = 0
if n<0:
    print('Hãy nhập vào số nguyên dương')
else:
    while n>0:
        cs = n%10
        if cs%2==0:
            chan+=1
        else:
            le += 1
        n//=10
    print(f'có {chan} số chẵn, {le} số lẻ')
```

# Bài 18: 
Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay 
không 

```python
n = int(input())
a=0
b=0
if n<=0:
    print('Nhập vào số nguyên dương')
else:
    while n>1:
        if n%3==0:
            a+=1
        else:
            b+=1
        n//=3
    if b!=0:
        print('Khong là số dạng 3^k')
    else:
        print('Là số dạng 3^k')
```

# Bài 19:
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số 
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong 
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
A=int(input())
print(f'A = {A}')
f1=1
f2=1
while f2<=A:
    f1t=f1
    f2t=f2
    f1=f1+f2
    f2=f1+f2
if f1<=A<=f2:
    print(f'So nho nhat la: f1={f1}')
elif f1>A:
    print(f'So nho nhat la f2t={f2t}')
```

# Bài 20:
Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = int(input())
dem=0
temp=n
while temp>0:
    cs=temp%10
    if cs in [2,3,5,7]:
        dem+=1
    temp//=10
print(f'{n} co {dem} so nguyen to')
```
