# W3A1 
Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ 
NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử 
lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý. 

```python
n = int(input())
if 100<=n<=999:
    temp = n
    s = 0
    while temp>0:
        a = temp%10
        s = s*10 + a
        temp//=10
    print(s)
else:
    print('Hãy nhập số nguyên dương có 3 chữ số')
```

# W3A11 
[Phân loại tam giác] 
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm 
tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in 
ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", 
"Tam giác cân", hay "Tam giác thường".

```python
a,b,c = map(int, input().split())
if a<b+c and b<a+c and c<a+b:
    if a==b==c:
        print('Tam giác đều')
    elif a==b or a==c or b==c:
        print('Tam giác cân')
    else:
        print('Tam giác thường')
else:
    print('Không phải tam giác')
```

# W3A16 
[Rounding] 
Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số 
nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả 
cách nhau một dấu cách

```python
a = float(input())
tp = a - int(a)
if a>=0:
    if a%1==0:
        print(a)
    else:
        if tp<0.5:
            nguyen = int(a)
        else:
            nguyen = int(a) + 1
        print(f'{int(a)+1} {int(a)} {nguyen}')
else:
    if a%1==0:
        print(a)
    else:
        if tp<0.5:
            nguyen = int(a)+1
        else:
            nguyen = int(a)
        print(f'{int(a)} {int(a)-1} {nguyen}')
```
