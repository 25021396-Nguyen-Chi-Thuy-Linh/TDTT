Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 

```python
def is_amstrong(x):
    
    # Một số nguyên dương được gọi là số Armstrong 
    # nếu tổng các chữ số của nó, sau khi lũy thừa với số mũ 
    # bằng chính số lượng chữ số của số đó, bằng chính nó.
    
    mu = len(str(x))
    s = 0
    temp = x
    while temp > 0:
        a = temp % 10
        s = s + a**mu
        temp //= 10
    return s == x

x = int(input())
print(is_amstrong(x))
```

# W5A19 
VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên. 


```python
# Bạn tính tổng các chữ số của một số. Nếu kết quả ra một số có nhiều hơn 1 chữ số, bạn lại tiếp tục cộng các chữ số đó lại cho đến khi chỉ còn lại 1 chữ số duy nhất. Nếu chữ số cuối cùng đó là 1, thì số ban đầu là số kỳ diệu.

def so_ky_dieu(n):
    s = 0
    temp = n
    while temp>0:
        a = temp%10
        s = s + a
        temp //= 10
    
    while s>=10:
        res = 0
        while s>0:
            res = res + s%10
            s //= 10
        s = res
 
    return s == 1

n = int(input())
print(so_ky_dieu(n))
```

# W5A25 
Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay 
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), 
(0,3)], p = (2,1) 

Giả sử tứ giác của bạn có 4 đỉnh theo thứ tự vòng tròn là $A, B, C, D$.Khi một điểm $P$ nằm hoàn toàn bên trong tứ giác, nếu bạn đi một vòng theo các cạnh $AB \rightarrow BC \rightarrow CD \rightarrow DA$, điểm $P$ sẽ luôn luôn nằm về cùng một phía (hoặc tất cả cùng bên trái, hoặc tất cả cùng bên phải) so với các đường thẳng chứa cạnh đó.Trong lập trình, để biết điểm $P$ nằm bên trái hay bên phải vectơ cạnh $\vec{AB}$, ta tính tích hướng (tích chéo) của hai vectơ $\vec{AB}$ và $\vec{AP}$.$$\text{Tích chéo} = (x_B - x_A) \times (y_P - y_A) - (y_B - y_A) \times (x_P - x_A)$$Nếu điểm $P$ nằm trên biên (cạnh của tứ giác), tích chéo của cạnh đó với điểm $P$ sẽ bằng đúng 0.Do đó, điều kiện để điểm $P$ nằm trong hoặc trên biên của tứ giác lồi là: Tất cả các tích chéo qua 4 cạnh phải cùng dấu ($\ge 0$) hoặc cùng dấu ($\le 0$).

```python
def point_in_quadrilateral(quad, p):
    xp, yp = p
    signs = []
    
    for i in range(4):
        xa, ya = quad[i]
        xb, yb = quad[(i+1) % 4]
        
        cross_prod = (xb - xa) * (yp-ya) - (yb - ya) * (xp - xa)
        
        signs.append(cross_prod)
        
    all_pos_or_zero = all(sign >= 0 for sign in signs)
    all_neg_or_zero = all(sign <= 0 for sign in signs)
    
    return all_pos_or_zero or all_neg_or_zero

quad = [(0, 0), (4, 0), (4, 3), (0, 3)]
p = (2, 1)

print(point_in_quadrilateral(quad, p))
```

# W5A28 
Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3). 


```python
import math
def goc(x1,y1,x2,y2,x3,y3):
    xba = x1 - x2
    yba = y1 - y2
    
    xbc = x3 - x2
    ybc = y3 - y2
    
    dot_prod = xba*xbc + yba*ybc
    
    len_ba = (xba**2 + yba**2)**0.5
    len_bc = (xbc**2 + ybc**2)**0.5
    
    if len_ba == 0 or len_bc == 0:
        return 0
    
    cos_abc = dot_prod / (len_ba * len_bc)
    
    cos_abc = max(-1.0, min(1.0, cos_abc))
    
    rad = math.acos(cos_abc)
    deg = math.degrees(rad)
    
    return deg

x1, y1, x2, y2, x3, y3 = map(float, input().split())
print(goc(x1,y1,x2,y2,x3,y3))
```

Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].

# W5A30 
Viết hàm tính hệ số nhị thức C(n, k). 


```python
def giai_thua(a):
    gt = 1
    for i in range(1, a+1):
        gt *= i
    return gt

def he_so(n,k):
    if k<0 or k>n:
        return 0
    
    tu = giai_thua(n)
    mau = giai_thua(k) * giai_thua(n-k)
    
    return tu // mau

n,k = map(int, input().split())
print(he_so(n,k))
```

# W5A34
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 
Nếu n ≤ 0 trả về 0.

```python
def chu_so_doi_mot_khac_nhau(x):
    chuoi_so  = str(x)
    
    return len(chuoi_so) == len(set(chuoi_so))

def sum_1_to_n(n):
    if n<=0:
        return 0
    
    s = 0
    for x in range(1, n+1):
        if chu_so_doi_mot_khac_nhau(x):
            s += x
            
        return s
    
n = int(input())
print(sum_1_to_n(n))
```
