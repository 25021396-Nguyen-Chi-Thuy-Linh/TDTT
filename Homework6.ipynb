# W6A16 
Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma 
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải 

```python
n, m = map(int,input().split())
matrix = []
for i in range(n):
    row = list(map(int,input().split()))
    matrix.append(row)
print(matrix)
for row in matrix:
    for i in row:
        print(f'{i:4d}', end='')
    print()
```

# W6A17 
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và 
đường chéo phụ của mảng 

```python
n = int(input())
mat = []
for i in range(n):
    row = list(map(int,input().split()))
    mat.append(row)
for row in mat:
    for i in row:
        print(f'{i}', end=' ')
    print()

chinh = []
phu = []

for i in range(n):
    chinh.append(mat[i][i])
    phu.append(mat[i][n-i-1])

print(f'{chinh}\n{phu}')
```

Nhập vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên 
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột

# W6A19 
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất 
trong list 

```python
def lm(ds):
    ln = 0
    for i in range(1, len(ds)):
        if ds[i] > ds[ln]:
            ln = i
    
    return ln
ds = list(map(int,input().split()))
print(lm(ds))
```

# W6A20 
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và 
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1

```python
ds = list(map(int, input().split()))
k = int(input())

def vitri(ds, k):
    for i in range(len(ds)):
        if ds[i] == k:
            return i
        
    return -1

print(vitri(ds, k))
```

# W6A21 
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ 
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

# W6A23

# W6A24 
Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí 
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9 

```python
def day(lst):
    lm0 = []
    lm = []
    for i in lst:
        if i == 0:
            lm0.append(i)
        else:
            lm.append(i)
    
    return lm0 + lm

lst = list(map(int, input().split()))
print(*day(lst))
```

# W6A27 
Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). 
nếu không có in ra -1 

```python
ds = list(map(int,input().split()))
x = int(input())

def vitri(ds,x):
    vt = []
    for i in range(len(ds)):
        if ds[i] < x:
            vt.append(i)
    
    if vt == []:
        return -1
    
    return vt

print(*vitri(ds,x))
```

# W6A28 
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm 
thực sự, hay vô danh (không tăng, không giảm) 

```python
ds = list(map(int, input().split()))
def tanggiamvodanh(ds):
    day_t = [ds[0]]
    day_g = [ds[0]]
    for i in range(1,len(ds)):
        if ds[i] > ds[i-1]:
            day_t.append(ds[i])
        elif ds[i] < ds[i-1]:
            day_g.append(ds[i])
    
    if len(day_t) == len(ds):
        return 'Day tang'
    
    if len(day_g) == len(ds):
        return 'Day giam'
    
    if len(day_t) != len(ds) and len(day_g) != len(ds):
        return 'Vo danh'
    
print(tanggiamvodanh(ds))
```
